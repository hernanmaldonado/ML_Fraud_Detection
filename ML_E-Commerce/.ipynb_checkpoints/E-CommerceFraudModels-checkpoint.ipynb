# <span style="color:#7E3517">Machine Learning Models for Fraud Detection</span>

This project is presented thourgh an Ipython notebook where I tune multiple machine learning models for fraud detection. The data used was obtained from Kaggle fraudulent e-commerce [Dataset](https://www.kaggle.com/datasets/shriyashjagtap/fraudulent-e-commerce-transactions).  

>"This synthetic dataset, "Fraudulent E-Commerce Transactions," is designed to simulate transaction data from an e-commerce platform with a focus on fraud detection. It contains a variety of features commonly found in transactional data, with additional attributes specifically engineered to support the development and testing of fraud detection algorithms."

Approaches taken on the [models](https://www.kaggle.com/datasets/shriyashjagtap/fraudulent-e-commerce-transactions/code) tried by other Kaggle engineers aim for a high accuracy. This is not a good approach because Fraud is typically an imbalanced data problem. Most corporations are more concerned about preventing fraud at the expense of a handful of false positives, this means that they are willing to trade precision for recall. To achieve this ML engineer must focus its attention on maximizing recall and specifically on finding multiple levels of potential fraud. 

As I will show on this notebook, The KS statistic helps us determine the decile (or bucket) that is most appropriate to separate fruad vs non-fraud but will also allow us to decide if, as a business, we want to put different lebels on suspicious activity rather than label it all equally. Multiple labels allow workflows to take different actions on transactions depending on the assigned level of fraud suspicion.

# <span style="color:green">Histograms</span>
The below function and code create histograms showing that fraud is more concentrated on: 
- Transactions with lower amounts
- Newer accounts
- In some age groups

These characteristics should help tree based models to separate fraudulent transactions from legitimate ones 

```python
def featurehist(data,feature,thisgraph):
    data[data['Is Fraudulent']==0][feature].plot(kind='hist',alpha=0.5,  label='No fraud',ax=thisgraph, legend=True)
    data[data['Is Fraudulent']==1][feature].plot(kind='hist',alpha=0.5 ,secondary_y=True, label='Fraud',ax=thisgraph, legend=True)
    thisgraph.set_ylabel('No Fraud')
    thisgraph.right_ax.set_ylabel('Fraud')
    thisgraph.set(xlabel=feature)
    return thisgraph
```

```python
fig, axes = plt.subplots(2, 2,figsize=(8, 8))
fig.tight_layout()
fig.subplots_adjust(wspace=0.8, hspace=0.5)
position=it.product(range(2),range(2))
features=['Customer Age','Account Age Days','Transaction Amount','Quantity']
for feature,pos in list(zip(features,position)):
        thisgraph=axes[pos[0],pos[1]]
        thisgraph=featurehist(train_input, feature,thisgraph)
plt.show()
```

# <span style="color:green">Graphs Fraud Percentage</span>
The below function and code create graphs that show percentage of fraud by different metrics. Higher levels mean that on that specific sub metric there is more fraud and its an indicator of a potentially good feature to include in the modelling work.

- Between 12:00 AM and 5:00 AM (Night time) 
- On younger and elder customers (probably identity theft from family memebers)
- On new accounts

These characteristics should help tree based models to separate fraudulent transactions from legitimate ones.

```python
def featurepct(data,feature,thisgraph):
    datasum=data.groupby(feature)['Is Fraudulent'].agg(['sum', 'count'])
    datasum['fraud_pct_hour']=datasum['sum'].div(datasum['count'])
    datasum[feature]=datasum.index
    datasum.plot.bar(x=feature, y='fraud_pct_hour', rot=0, ax=thisgraph, legend=False)
    thisgraph.set_ylabel('Fraud %')
    thisgraph.set(xlabel=feature)
    return thisgraph
```

```python
def plotpct(data:pd.DataFrame, features:list, numrows:int):
    fig, axes = plt.subplots(numrows, 2,figsize=(12, 12))
    fig.tight_layout()
    fig.subplots_adjust(wspace=0.3, hspace=0.3)
    position=it.product(range(numrows),range(2))
    if len(features)%2==1:
        axes[int((round(len(features))/2)),1].set_visible(False)
    for feature,pos in list(zip(features,position)):
            thisgraph=axes[pos[0],pos[1]]
            thisgraph=featurepct(data, feature,thisgraph)
    plt.show()
features=['Payment Method', 'Product Category', 'Device Used', 'Transaction Hour', 'Customer Age','Account Age Days','Quantity']
plotpct(train_input,features,4)
```

## <span style="color:green">Common fraud practices</span>
### We will explore some common ways to catch or at least curb fraud practices

One common way to flag a transaction for potential fraud is if the billing address, zip or state is different than the shipping address, zip or state. This happens because people steal information from customers online but they aren't phuysically located on the same area. The below graphs explore that but conclude that in this dataset there is no indication of that type of fraud occuring prevalently.   

Another common way to flag potential transactions as fraudulent is if the come from a state or zip that has shown high proportions of fraud when compared to regular transactions. In this dataset we see that there are some states (billing and shipping) with higher concentrations of fraud. However when we drill down to the zip level the concentration is more marked in some zips. This should help the model use the zips as a potential flag for fraud.

IP addresses can be used multiple times by fraudsters that are not very sophisticated. Ideally, those IPs  are flagged and black listed such that any transaction comming from those IP addresses gets either blocked or carefully processed. However, in this dataset we only see one repeating IP address with fraud.

## <span style="color:green">Random Forest</span>
Below is the Random Forest training and testing code that was fine tuned to maximize recall. As explained before, recall is for most business the optimal metric when searching for fraudulent transactions, as preventing fraud is usually the highest priority, in general it is admissible to flag somenon fraud transactions as potential fraud and later clearing them as not fraud. 

The results are encouraging as we see a 0.85 recall and ks of 47.3 with the higher probability buckets accumulating mostly fraud. This result allows us to separate model results by labeling as fraud (high probability buckets), potential fraud (medium high probability buckets) and no fraud (low probability buckets).  

```python
continous=Pipeline(steps=[('scaler',StandardScaler())])
categorical=Pipeline(steps=[('onehot',OneHotEncoder(handle_unknown='ignore'))])
#pipe = Pipeline(steps=[('clf', RandomForestClassifier())])
preprocesser=ColumnTransformer(transformers=[('num',continous,numer),('cat',categorical,categ)],remainder='passthrough')
clf = RandomForestClassifier()
pipeline=Pipeline(steps=[('preprocessor',preprocesser),('randF',clf)])
```

## <span style="color:green">Random Forest Results</span>

As mentioned before, the results are encouraging as we see a 0.85 recall. However it is important to see if there is good probability separation in oirder to crete buckets with duffierent fraud labels. We will explore that below with the KS. 

```python
#taken from https://www.listendata.com/2019/07/KS-Statistics-Python.html
def ks(data=None,target=None, prob=None):
    data['target0'] = 1 - data[target]
    data['bucket'] = pd.qcut(data[prob], 10)
    grouped = data.groupby('bucket', as_index = False, observed=False)
    kstable = pd.DataFrame()
    kstable['min_prob'] = grouped.min()[prob]
    kstable['max_prob'] = grouped.max()[prob]
    kstable['events']   = grouped.sum()[target]
    kstable['nonevents'] = grouped.sum()['target0']
    kstable = kstable.sort_values(by="min_prob", ascending=False).reset_index(drop = True)
    kstable['event_rate'] = (kstable.events / data[target].sum()).apply('{0:.2%}'.format)
    kstable['nonevent_rate'] = (kstable.nonevents / data['target0'].sum()).apply('{0:.2%}'.format)
    kstable['cum_eventrate']=(kstable.events / data[target].sum()).cumsum()
    kstable['cum_noneventrate']=(kstable.nonevents / data['target0'].sum()).cumsum()
    kstable['KS'] = np.round(kstable['cum_eventrate']-kstable['cum_noneventrate'], 3) * 100

    #Formating
    kstable['cum_eventrate']= kstable['cum_eventrate'].apply('{0:.2%}'.format)
    kstable['cum_noneventrate']= kstable['cum_noneventrate'].apply('{0:.2%}'.format)
    kstable.index = range(1,11)
    kstable.index.rename('Decile', inplace=True)
    pd.set_option('display.max_columns', 9)
    print(kstable)
    
    #Display KS
    from colorama import Fore
    print(Fore.RED + "KS is " + str(max(kstable['KS']))+"%"+ " at decile " + str((kstable.index[kstable['KS']==max(kstable['KS'])][0])))
    return(kstable)
```

## <span style="color:green">KS analysis</span>

For probabilities over 0.5 we get a 47.3 KS and we capture 85% of fraud and we label about 37% of good transactions as either fraud or potential fraud. If we go tot he second bucket we get 42.14% of fraud and only 4.81% of not fraud.

```python
def modks(data=None,target=None, prob=None, buckets=None):
    data['target0'] = 1 - data[target]
    #data['bucket'] = pd.qcut(data[prob], 10)
    data[['bucket_num','bucket']]=data.apply(lambda x: bucketmap(x[prob],buckets),axis=1,result_type='expand')
    grouped = data.groupby('bucket', as_index = False)
    kstable = pd.DataFrame()
    kstable['min_prob'] = grouped.min()[prob]
    kstable['max_prob'] = grouped.max()[prob]
    kstable['events']   = grouped.sum()[target]
    kstable['nonevents'] = grouped.sum()['target0']
    kstable = kstable.sort_values(by="min_prob", ascending=False).reset_index(drop = True)
    kstable['event_rate'] = (kstable.events / data[target].sum()).apply('{0:.2%}'.format)
    kstable['nonevent_rate'] = (kstable.nonevents / data['target0'].sum()).apply('{0:.2%}'.format)
    kstable['cum_eventrate']=(kstable.events / data[target].sum()).cumsum()
    kstable['cum_noneventrate']=(kstable.nonevents / data['target0'].sum()).cumsum()
    kstable['KS'] = np.round(kstable['cum_eventrate']-kstable['cum_noneventrate'], 3) * 100

    #Formating
    kstable['cum_eventrate']= kstable['cum_eventrate'].apply('{0:.2%}'.format)
    kstable['cum_noneventrate']= kstable['cum_noneventrate'].apply('{0:.2%}'.format)
    
    kstable.index = range(1,len(kstable)+1)
    kstable.index.rename('Bucket', inplace=True)
    pd.set_option('display.max_columns', 9)
    print(kstable)
    
    #Display KS
    from colorama import Fore
    print(Fore.RED + "KS is " + str(max(kstable['KS']))+"%"+ " at bucket " + str((kstable.index[kstable['KS']==max(kstable['KS'])][0])))
    return(kstable)
```

# <span style="color:green">Isolation Forest</span>

Below is the Isolation Forest training and testing code, it is an unsupervised machine learning model designed for outlier detection. 
The results are not encouraging as we see the model get a really low recall (.1) making it not effective for our fraud detection purposes.

<span style="color:red">**Important:**</span> For interpretation purposes note that Isolation Forest labels outliers as -1 and normal observations as 1. 

```python
param_grid_iso = {	'isoF__n_estimators': [5,10,50,100,500], #The number of base estimators in the ensemble.
				'isoF__max_samples': [5,10,50,500], #The number of samples to draw from X to train each base estimator.
				'isoF__contamination': [0.1], #The amount of contamination of the data set, i.e. the proportion of outliers in the data set. Used when fitting to define the threshold on the scores of the samples.
				'isoF__max_features': [2,4,6,10], #The number of features to draw from X to train each base estimator.
				'isoF__bootstrap': [True, False], #f True, individual trees are fit on random subsets of the training data sampled with replacement. If False, sampling without replacement is performed.
				'isoF__n_jobs': [-1], #The number of jobs to run in parallel for both fit and predict. None means 1 unless in a joblib.parallel_backend context
				'isoF__random_state':[42], #Controls the pseudo-randomness of the selection of the feature and split values for each branching step and each tree in the forest.
				'isoF__verbose':[2], #Controls the verbosity of the tree building process
				'isoF__warm_start':[False] #When set to True, reuse the solution of the previous call to fit and add more estimators to the ensemble, otherwise, just fit a whole new forest.
             }
```

```python
iso_y=y.copy()
iso_y.replace(to_replace = 1, value = -1,inplace=True)
iso_y.replace(to_replace = 0, value = 1,inplace=True)
```

```python
preprocesser=ColumnTransformer(transformers=[('num',continous,numer),('cat',categorical,categ)],remainder='passthrough')
clf_iso = IsolationForest()
pipeline_iso=Pipeline(steps=[('preprocessor',preprocesser),('isoF',clf_iso)])

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
iso_random = RandomizedSearchCV(pipeline_iso,                      
                                scoring = "recall", 
                               refit=True,
                               return_train_score = True,  
                               param_distributions = param_grid_iso, 
                               n_iter = 100, cv = cv, 
                               verbose=2, 
                               random_state=42, 
                               n_jobs = -1)
iso_random.fit(X,iso_y)

import joblib
joblib.dump(iso_random.best_estimator_, 'iso_random.pkl')
```

```python
y_pred_iso=iso_random.predict(X_test)
```

As mentioned before, the results are encouraging as we see a 0.80 recall vs the 0.85 recall of the random forest. A KS of 49.9 higher than the 47.3 of the Random forest and more importantly a much more nuanced decile distribution that makes it easier to use multiple labels when flagging fraud. 



```python
import optuna
import xgboost as xgb
```

```python
def objective(trial,X=X,y=y,X_test=X_test,y_test=y_test):
    param = {
        'tree_method':'hist',  # this parameter means using the GPU when training our model to speedup the training process
        'device':'cuda',
        'lambda': trial.suggest_float('lambda', 1e-3, 15.0),
        'alpha': trial.suggest_float('alpha', 1e-3, 15.0),
        'colsample_bytree': trial.suggest_categorical('colsample_bytree', [0.3,0.4,0.5,0.6,0.7,0.8,0.9, 1.0]),
        'subsample': trial.suggest_categorical('subsample', [0.4,0.5,0.6,0.7,0.8,1.0]),
        'learning_rate': trial.suggest_categorical('learning_rate', [0.008,0.01,0.012,0.014,0.016,0.018, 0.02]),
        'n_estimators': trial.suggest_categorical('n_estimators', [50,75,100,125,150,200,250]),
        'max_depth': trial.suggest_int('max_depth', 5,30),
        'random_state': trial.suggest_categorical('random_state', [42]),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 300),
        'scale_pos_weight':trial.suggest_int('scale_pos_weight', 10, 50),
        'enable_categorical':[True],
        'eval_metric':['error']
    }
    preprocesser=ColumnTransformer(transformers=[('num',continous,numer),('cat',categorical,categ)],remainder='passthrough')
    model = xgb.XGBClassifier(**param)
    pipeline=Pipeline(steps=[('preprocessor',preprocesser),('model',model)])
      
    
    pipeline.fit(X,y)
    
    preds = pipeline.predict(X_test)
    
    recall = recall_score(y_test, preds, average='macro')
    
    return recall
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=30)
print('Number of finished trials:', len(study.trials))
print('Best trial:', study.best_trial.params)
```

```python
best_trial_parms = study.best_trial.params
preprocesser=ColumnTransformer(transformers=[('num',continous,numer),('cat',categorical,categ)],remainder='passthrough')
model = xgb.XGBClassifier(**best_trial_parms)
pipeline=Pipeline(steps=[('preprocessor',preprocesser),('model',model)])
pipeline.fit(X,y)
#model = instantiate_model(best_trial, numerical_columns, categorical_columns)
#model.fit(X_train, y_train)
```

```python
preds = pipeline.predict(X_test)
y_pred=pipeline.predict(X_test)
y_pred_prob=pipeline.predict_proba(X_test)
recall = recall_score(y_test, preds)
print("Recall is:" +str(recall))
```

```python
def roc_plot(y_true, y_probs, label, compare=False, ax=None):
    """ Plot Receiver Operating Characteristic (ROC) curve 
        Set `compare=True` to use this function to compare classifiers. """
    
    fpr, tpr, thresh = roc_curve(y_true, y_probs,
                                 drop_intermediate=False)
    auc = round(roc_auc_score(y_true, y_probs), 2)
    
    fig, axis = (None, ax) if ax else plt.subplots(nrows=1, ncols=1)
    label = ' '.join([label, f'({auc})']) if compare else None
    sns.lineplot(x=fpr, y=tpr, ax=axis, label=label)
    
    if compare:
        axis.legend(title='Classifier (AUC)', loc='lower right')
    else:
        axis.text(0.72, 0.05, f'AUC = { auc }', fontsize=12,
                  bbox=dict(facecolor='green', alpha=0.4, pad=5))
            
        # Plot No-Info classifier
        axis.fill_between(fpr, fpr, tpr, alpha=0.3, edgecolor='g',
                          linestyle='--', linewidth=2)
        
    axis.set_xlim(0, 1)
    axis.set_ylim(0, 1)
    axis.set_title('ROC Curve')
    axis.set_xlabel('False Positive Rate [FPR]\n(1 - Specificity)')
    axis.set_ylabel('True Positive Rate [TPR]\n(Sensitivity or Recall)')
    
    plt.close()
    
    return axis if ax else fig
```

```python
import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
roc_plot(y_test, y_pred_prob[:,1], label=None)
```
